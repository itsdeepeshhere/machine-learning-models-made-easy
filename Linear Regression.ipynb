```python
import pandas as pd
```

```python
dataset=pd.read_csv('iris.csv')
dataset.head()
```

```python
dataset['sepal_width']>3
```

```python
dataset[dataset['sepal_width']>4]
```

## Linear Regression-

```python
x=dataset[['sepal_width']]
y=dataset[['sepal_length']]
```

```python
from sklearn.model_selection import train_test_split #used in every ML model to split data into training and testing dataset
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
#importing linear regression model-
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
#comparing y_test and y_pred
y_pred[:5]
```

```python
y_test.head()
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,y_pred)
```

## Multiple Linear Regression-

```python
x1=dataset[['sepal_width','petal_length','petal_width']]
y1=dataset[['sepal_length']]
```

```python
x_train1,x_test1,y_train1,y_test1=train_test_split(x1,y1,test_size=0.3)
```

```python
lr2=LinearRegression()
lr2.fit(x_train1,y_train1)
```

```python
y_pred2=lr2.predict(x_test1)
```

```python
mean_squared_error(y_pred2,y_test1)
```
